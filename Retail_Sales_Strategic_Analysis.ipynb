```python


import matplotlib.pyplot as plt
import numpy as np


months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
months_sale= np.random.randint(50,250, size=12)
product=["Sugar","Flour","Oil","Salt"]
product_sale=[350,500,430,250]
region=["Punjab","Sindh","Balochistan","Khyber Pakhtun khaun"]
region_share=[25,35,30,10]
orders=np.random.normal(125,73,250)
fig,axs =plt.subplots(2,2,figsize=(13,10))
axs[0,0].plot(months,months_sale,color="#4f1b1d",marker="^")
axs[0,0].set_xlabel("Months")
axs[0,0].set_ylabel("Sales($)")
axs[0,0].set_title("Monthly Sales Comparision",bbox=dict(facecolor='pink', alpha=0.5, ))
axs[0,0].grid(True,alpha=0.3)
max_month=months[np.argmax(months_sale)]
max_sales=np.max(months_sale)
axs[0,0].annotate(f"Peak:{max_sales}",fontsize=9,
                  xy=(max_month,max_sales),
                  xytext=(max_month,max_sales+17),arrowprops=dict(facecolor="green",shrink=0.0005),ha="center")

min_month=months[np.argmin(months_sale)]
min_sales=np.min(months_sale)
axs[0,0].annotate(f"Lowest:{min_sales}",fontsize=9,
                  xy=(min_month,min_sales),
                  xytext=(min_month,min_sales+20),arrowprops=dict(facecolor="red",shrink=0.0005),ha="center")

axs[0,1].bar(product,product_sale,color=["#5c4f4a","#425438","#385445","#513854"])
axs[0,1].set_xlabel("Products")
axs[0,1].set_ylabel("Product_Sales")
axs[0,1].set_title("Sales by Products",bbox=dict(facecolor='pink', alpha=0.5, ))
axs[0,1].grid(True,alpha=0.3)

best_index = np.argmax(product_sale)
best_product = product[np.argmax(product_sale)]
best_value = np.max(product_sale)
axs[0, 1].text(best_index, best_value+7, f"Top Product: {best_product}", ha="center", color="green")

wrost_index = np.argmin(product_sale)
best_product = product[np.argmin(product_sale)]
best_value = np.min(product_sale)
axs[0, 1].text(wrost_index, best_value+7, f"Wrost Product: {best_product}", ha="center", color="red")

axs[1,0].pie(region_share,labels=region,autopct="%1.1f%%",shadow=True,startangle=90,explode=[0,0.1,0,0],colors=["#6b665c","#edb2bc","#7a7341","#84baae"])
axs[1,0].set_title("Market Share BY Region",bbox=dict(facecolor='pink', alpha=0.5, ))

axs[1,1].hist(orders,bins=15,color="#c99c89",edgecolor="black")
axs[1,1].set_xlabel("Order Price")
axs[1,1].set_ylabel("Order Quantity")
axs[1,1].set_title("Order's Distribution",bbox=dict(facecolor='pink', alpha=0.5))
axs[1,1].grid(True,alpha=0.3)
fig.suptitle("Company Sales Dashboard(Analyst report)",bbox=dict(facecolor='#9c966b', alpha=0.5),fontsize=17,fontweight="bold")
fig.text(0.5, 0.01,
         "Insights:", ha="center", fontsize=10, fontweight="bold",bbox=dict(facecolor='pink', alpha=0.5))
fig.text(0.5, -0.06,
         f"~ Monthly sales peaked in {max_month} with ${max_sales} revenue.\n"
         f"~ Product {best_product} is the best-selling item.\n"
         "~Sindh region leads market share with 35%.\n"
         "~Most orders fall in the $150â€“250 range.",
         ha="center",fontsize=10)
```
